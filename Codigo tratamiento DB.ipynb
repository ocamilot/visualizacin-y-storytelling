```python
import pandas as pd
import requests

# Diccionario de monedas principales
currency_map_main = {
    "USD": "USD", "GBP": "GBP", "CAD": "CAD", "EUR": "EUR", "AUD/NZD": "AUD",
    "CHF": "CHF", "ZAR": "ZAR", "SEK": "SEK", "HKD": "HKD", "JPY": "JPY"
}

# Diccionario de monedas "Other"
currency_map_other = {
    "INR": "INR", "Peso Argentino": "ARS", "MYR": "MYR", "CHF": "CHF", "NOK": "NOK",
    "USD": "USD", "BR$": "BRL", "SEK": "SEK", "Dkk": "DKK", "EUR": "EUR", "TTD": "TTD",
    "Indian rupees": "INR", "BRL (R$)": "BRL", "Mexican pesos": "MXN", "CZK": "CZK",
    "GBP": "GBP", "Bdt": "BDT", "ZAR": "ZAR", "American Dollars": "USD", "Php": "PHP",
    "PLN (Polish zloty)": "PLN", "czech crowns": "CZK", "TRY": "TRY",
    "Norwegian kroner (NOK)": "NOK", "CNY": "CNY", "ILS/NIS": "ILS", "US Dollar": "USD",
    "BRL": "BRL", "NIS (new Israeli shekel)": "ILS", "RMB (chinese yuan)": "CNY",
    "AUD": "AUD", "Taiwanese dollars": "TWD", "Philippine Peso": "PHP", "SGD": "SGD",
    "KRW (Korean Won)": "KRW", "THB": "THB", "IDR": "IDR", "NZD": "NZD", "ILS (Shekel)": "ILS",
    "DKK": "DKK", "China RMB": "CNY", "AUD Australian": "AUD", "LKR": "LKR",
    "Polish Z≈Çoty": "PLN", "Philippine peso (PHP)": "PHP", "Australian Dollars": "AUD",
    "ARS": "ARS", "Argentinian peso (ARS)": "ARS", "Israeli Shekels": "ILS", "ILS": "ILS",
    "MXN": "MXN", "PhP (Philippine Peso)": "PHP", "PLN": "PLN", "KRW": "KRW",
    "SAR": "SAR", "RM": "MYR", "Argentine Peso": "ARS", "Philippine Pesos": "PHP",
    "INR (Indian Rupee)": "INR", "NTD": "TWD", "Danish Kroner": "DKK", "CAD": "CAD",
    "Korean Won": "KRW", "Euro": "EUR", "THAI BAHT": "THB", "Thai Baht": "THB",
    "croatian kuna": "HRK", "PLN (Zwoty)": "PLN", "Rupees": "INR", "Singapore Dollara": "SGD",
    "NGN": "NGN", "COP": "COP", "TZS": "TZS", "Pesos Colombianos": "COP"
}

# Función para obtener tasas de cambio desde la API
def obtener_tasas_de_cambio():
    url = "https://api.exchangerate-api.com/v4/latest/USD"  # Base USD
    response = requests.get(url)
    data = response.json()
    return data["rates"]  # Retorna un diccionario de tasas de cambio

# Obtener las tasas de cambio una sola vez
#exchange_rates = obtener_tasas_de_cambio()

exchange_rates = {
    "USD": 4130.87,  # Dólar estadounidense
    "GBP": 5083.75,  # Libra esterlina
    "CAD": 3240.50,  # Dólar canadiense
    "EUR": 4520.50,  # Euro
    "AUD": 2900.30,  # Dólar australiano
    "NZD": 2700.20,  # Dólar neozelandés
    "CHF": 4200.10,  # Franco suizo
    "ZAR": 280.15,   # Rand sudafricano
    "SEK": 450.25,   # Corona sueca
    "HKD": 530.40,   # Dólar de Hong Kong
    "JPY": 36.50,    # Yen japonés
    "INR": 50.30,    # Rupia india
    "ARS": 25.60,    # Peso argentino
    "MYR": 980.45,   # Ringgit malayo
    "NOK": 420.35,   # Corona noruega
    "BRL": 800.75,   # Real brasileño
    "DKK": 605.20,   # Corona danesa
    "TTD": 610.50,   # Dólar de Trinidad y Tobago
    "MXN": 210.80,   # Peso mexicano
    "CZK": 190.90,   # Corona checa
    "BDT": 48.60,    # Taka bangladesí
    "PHP": 74.20,    # Peso filipino
    "PLN": 1000.00,  # Zloty polaco
    "TRY": 550.75,   # Lira turca
    "CNY": 640.50,   # Yuan chino
    "ILS": 1200.30,  # Nuevo shéquel israelí
    "TWD": 150.40,   # Nuevo dólar taiwanés
    "SGD": 3050.60,  # Dólar de Singapur
    "KRW": 3.50,     # Won surcoreano
    "THB": 125.70,   # Baht tailandés
    "IDR": 0.29,     # Rupia indonesia
    "LKR": 22.50,    # Rupia de Sri Lanka
    "SAR": 1100.80,  # Riyal saudí
    "NGN": 10.50,    # Naira nigeriano
    "TZS": 1.78,     # Chelín tanzano
    "COP": 1.00      # Peso colombiano
}



# Agregar columnas de salario en COP
df["Annual_Salary_COP"] = 0
df["Compensation_COP"] = 0

# Convertir valores a COP
for index, row in df.iterrows():
    
    moneda = row["Currency"]
    salario = row["Annual_Salary"]
    salario = int(salario.replace(",", "")) 
    compensacion = row["Additional_Compensation"]
    
    # Validar si la moneda está en currency_map_main
    if moneda in currency_map_main:
        iso_code = currency_map_main[moneda]
    elif moneda == "Other":
        other_currency = row["If \"Other,\" please indicate the currency here: "]
        iso_code = currency_map_other.get(other_currency, "COP")
    else:
        iso_code = "COP"
    
    # Si la moneda es COP, mantener el valor original
    if iso_code == "COP":
        df.at[index, "Annual_Salary_COP"] = salario
        df.at[index, "Compensation_COP"] = compensacion
    else:
        tasa_cambio = exchange_rates.get(iso_code, 1)  # Obtener tasa de cambio o 1 si no está
        df.at[index, "Annual_Salary_COP"] = salario * tasa_cambio
        df.at[index, "Compensation_COP"] = compensacion * tasa_cambio
```
